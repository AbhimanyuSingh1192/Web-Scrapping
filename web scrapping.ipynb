```python
from bs4 import BeautifulSoup as bs
import pandas as pd
import requests
```

```python
page = requests.get("http://quotes.toscrape.com/page/2/")
page
```

```python
soup = bs(page.content) 
soup
```

```python
soup.find_all(class_='text')
```

```python
quotes = [i.text for i in soup.find_all(class_='text')]
quotes
```

```python
soup.find_all(class_='author')
```

```python
authors = [i.text for i in soup.find_all(class_='author')]
```

```python
authors
```

```python
df = pd.DataFrame()
df['Authors'] = authors
df['Quotes'] = quotes
```

```python
df
```

# more data

```python
urls=[f"http://quotes.toscrape.com/page/{i}/" for i in range(1,3)]
urls
```

```python
import time
for i in urls:
    page = requests.get(i)
    soup = bs(page.content) 
    quotes = [i.text for i in soup.find_all(class_='text')]
    authors = [i.text for i in soup.find_all(class_='author')]
    df = pd.DataFrame()
    df['Authors'] = authors
    df['Quotes'] = quotes
    display(df)
    time.sleep(3)
```
